```python
# Water Jug Problem using dfs
def d_fs(jg_1_cp, jg_2_cp, go_al, st, visted, pth, rules):
    x, y = st

    # If it is already visted then stop
    if st in visted:
        return False

    visted.add(st)
    pth.append(st)
    print("Curent state:", st)

    # Check goal
    if x == go_al or y == go_al:
        print("goal Reached")
        return True

    # Rules
    #1.can full Fill Jug 1
    #2.can Fill Jug 2
    #3.can Empty Jug 1
    #4.can Empty Jug 2
    #5.can Pour Jug 1 into Jug 2
    #6.can Pour Jug 2 into Jug 1

    if d_fs(jg_1_cp, jg_2_cp, go_al, (jg_1_cp, y), visted, pth, rules):
        rules.append("Filing the Jug 1")
        return True
    
    if d_fs(jg_1_cp, jg_2_cp, go_al, (x, jg_2_cp), visted, pth, rules):
        rules.append("Filig Jug 2")
        return True
    
    if d_fs(jg_1_cp, jg_2_cp, go_al, (0, y), visted, pth, rules):
        rules.append("Empty the Jug 1")
        return True
   
    if d_fs(jg_1_cp, jg_2_cp, go_al, (x, 0), visted, pth, rules):
        rules.append("Emptying the Jug 2")
        return True
    
    trnsfer = min(x, jg_2_cp - y)
    if d_fs(jg_1_cp, jg_2_cp, go_al, (x - trnsfer, y + trnsfer), visted, pth, rules):
        rules.append("Pouring Jug 1 into Jug 2")
        return True
   
    trnsfer = min(y, jg_1_cp - x)
    if d_fs(jg_1_cp, jg_2_cp, go_al, (x + trnsfer, y - trnsfer), visted, pth, rules):
        rules.append("Pouring Jug 2 into Jug 1")
        return True

    pth.pop() #back tracking
    return False


jug_1 = int(input("Enter cap of Jug 1: "))
jug_2 = int(input("Enter cap of Jug 2: "))
go_al = int(input("Enter goal amount: "))

visted_sts = set()
sol_pth = []
rules_used = []

print("jug 1 capacity",jug_1)
print("jug 2 capacity",jug_2)
print("goal",go_al)
print("\nStarting dfs\n")

found = d_fs(jug_1, jug_2, go_al, (0, 0), visted_sts, sol_pth, rules_used)

if found:
    print("\nRules Applied:")
    for r in reversed(rules_used):   # reverse to fix order
        print(r)

    print("\nSolution path:")
    for stp in sol_pth:
        print(stp)
else:
    print("\nNo Solution Found")
```
